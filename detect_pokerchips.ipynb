```python
# play video 
import cv2
import numpy as np
import os

# Playing video from file:
cap = cv2.VideoCapture('data/videos/She Is A SAVAGE at the Poker Table ♠️ PokerStars.mp4')

try:
    if not os.path.exists('data/images'):
        os.makedirs('data/images')

except OSError:
    print ('Error: Creating directory of data/images')

currentFrame = 0
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    print(ret)
    # Saves image of the current frame in jpg file
    name = 'data/images/frame' + str(currentFrame) + '.jpg'
    print ('Creating...' + name)
    cv2.imwrite(name, frame)

    # To stop duplicate images
    currentFrame += 1

```
