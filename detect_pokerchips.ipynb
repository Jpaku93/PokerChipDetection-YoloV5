```python
import cv2
from pytube import YouTube

# Set the YouTube video URL
video_url = "https://www.youtube.com/watch?v=Ts8Yh2DZA1Y"

# Download the YouTube video

# Download the YouTube video with the desired resolution
desired_resolution = "480p"  # Set the desired resolution, e.g., "720p", "1080p", "480p", etc.
# Download the YouTube video
yt = YouTube(video_url)
video = yt.streams.filter(res=desired_resolution).first()
path = video.download("videos")
```

```python
# Extract frames from the video, skipping every 20 frames
vidcap = cv2.VideoCapture(video.default_filename)
success, image = vidcap.read()
count = 0
frame_skip = 120  # Set the number of frames to skip

while success:
    if count % frame_skip == 0:
        # Save frame as JPEG file
        cv2.imwrite(f"frame{count}.jpg", image)

    success, image = vidcap.read()
    count += 1

# Release the video capture object
vidcap.release()
```
